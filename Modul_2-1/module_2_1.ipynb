## Daten aus Bibsonomy laden

Datenabfragen zu "Silvester" und "Neujahr"

```python
import requests
```

```python
url_neujahr = "https://www.bibsonomy.org/json/search/Neujahr?items=1000"
url_silvester = "https://www.bibsonomy.org/json/search/Silvester?items=1000"
```

```python
search_result = requests.get(url_neujahr)
json_neujahr = search_result.json()
#json_search_result
```

```python
search_result = requests.get(url_silvester)
json_silvester = search_result.json()
#json_search_result
```

## Daten in Pandas DateFrames laden und kombinieren

```python
import pandas as pd
```

```python
json_neujahr.keys()
```

```python
df_neujahr = pd.DataFrame(json_neujahr['items'])
df_neujahr['search_term'] = 'Neujahr'
#df_neujahr
```

```python
df_silvester = pd.DataFrame(json_silvester['items'])
df_silvester['search_term'] = "Silvester"
#df_silvester
```

```python
print(f"Silvestertabellengröße: {df_silvester.shape}\nNeujahrstabellengröße: {df_neujahr.shape}")
```

```python
df_jahreswechsel = pd.concat([df_silvester, df_neujahr])
df_jahreswechsel.shape
```

### Wenig genutzte Spalten herausfiltern

```python
# Methodentesten
leere_spalten = []
for key in df_jahreswechsel_select.keys():
    #print(df_jahreswechsel[key].count())
    if df_jahreswechsel_select[key].empty:
        leere_spalten.append(key)
```

```python
print(leere_spalten)
```

```python
wenig_genutzte_spalten = [column for column in df_jahreswechsel_select.keys() if df_jahreswechsel[column].count() < 500 ]
```

```python
print(wenig_genutzte_spalten)
```

```python
genutzte_spalten = [key for key in df_jahreswechsel_select.keys() if key not in wenig_genutzte_spalten]
print(genutzte_spalten)
```

```python
type(genutzte_spalten)
```

```python
df_jahreswechsel_viel_genutzt = df_jahreswechsel_select[genutzte_spalten]
```

```python
df_jahreswechsel_viel_genutzt.keys()
```

### Weitere Spalten löschen, Spalten umbenennen

```python
df_jahreswechsel_viel_genutzt = df_jahreswechsel_viel_genutzt.drop(columns=['type', 'count', 'intraHash', 'interHash', 'authors', 'editors'])
```

```python
df_jahreswechsel_viel_genutzt = df_jahreswechsel_viel_genutzt.rename(columns={'ee': 'pid', 'author': 'authors', 'editor': 'editors', 'label': 'title'})
df_jahreswechsel_viel_genutzt.drop_duplicates('title')
```

```python
df_jahreswechsel_viel_genutzt.keys()
```

### Zeilen ohne Autor und Titel löschen

```python
type(df_jahreswechsel_viel_genutzt.loc[df_jahreswechsel_viel_genutzt['authors'].notna()])
```

```python
# Frage: auch kombinierbar?? Problem: notna() gibt DataFrame aus, nicht Boolean
df_jahreswechsel_viel_und_voll = df_jahreswechsel_viel_genutzt.loc[df_jahreswechsel_viel_genutzt['authors'].notna()]
df_jahreswechsel_viel_und_voll = df_jahreswechsel_viel_und_voll.loc[df_jahreswechsel_viel_und_voll['title'].notna()]
```
