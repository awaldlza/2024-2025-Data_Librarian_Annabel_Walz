```python
import requests
```

```python
url_neujahr = "https://www.bibsonomy.org/json/search/Neujahr?items=1000"
url_silvester = "https://www.bibsonomy.org/json/search/Silvester?items=1000"
```

```python
search_result = requests.get(url_neujahr)
json_neujahr = search_result.json()
#json_search_result
```

```python
search_result = requests.get(url_silvester)
json_silvester = search_result.json()
#json_search_result
```

```python
import pandas as pd
```

```python
json_neujahr.keys()
```

```python
df_neujahr = pd.DataFrame(json_neujahr['items'])
df_neujahr['search_term'] = 'Neujahr'
#df_neujahr
```

```python
df_silvester = pd.DataFrame(json_silvester['items'])
df_silvester['search_term'] = "Silvester"
#df_silvester
```

```python
print(f"Silvestertabellengröße: {df_silvester.shape}\nNeujahrstabellengröße: {df_neujahr.shape}")
```

```python
df_jahreswechsel = pd.concat([df_silvester, df_neujahr])
df_jahreswechsel.shape
```
